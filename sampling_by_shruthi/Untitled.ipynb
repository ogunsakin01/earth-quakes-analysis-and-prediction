```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
eq_df = pd.read_csv(r'C:\Users\mishra\Downloads\eqdata.csv')
eq_df.isnull().sum()
```

```python
axes=eq_df.plot(kind='scatter',y='Depth', x='Magnitude',label='MEQ', color='green')
```

```python
eq_df.columns
feature_df=eq_df[[ 'Latitude', 'Longitude', 'Depth', 'Magnitude']]

#independent variables
X=np.asarray(feature_df)

#dependent variable
Y=np.asarray(eq_df['Depth'])
X[0:5]
```

```python
'''80:20 partition'''
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=5)

```

```python
from sklearn import svm
#gamma=co-efficient ,C=penalty
classifier = svm.SVC(kernel='linear',gamma='auto',C=2)
classifier.fit(X_train,Y_train)
Y_predict = classifier.predict(X_test)
```
